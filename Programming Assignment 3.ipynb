```python
import csv
import numpy as np
import pandas as pd
import networkx as nx
import gurobipy as gp
from gurobipy import GRB, Model
```

```python
df = pd.read_csv('hw3_graph.csv')
```

```python
numpy = df.to_numpy()
g = nx.DiGraph()

for each in range(df.shape[0]):
    g.add_edge(numpy[each, 0], numpy[each, 1], weight = numpy[each, 2])

path = nx.dijkstra_path(g, 0, 99)
```

```python
m = Model('Assignment 3')
m.setParam('OutputFlag', True)
```

```python
n = max(df.max().tolist()[:2]) + 1
x = m.addVars(n, n, name = 'x')
print('passed')
```

```python
m.addConstr(
    gp.quicksum(x[0, row['destination']] for index, row in df.iterrows() if row['origin'] == 0) == 1
)
m.addConstr(
    gp.quicksum(x[row['origin'], n-1] for index, row in df.iterrows() if row['destination'] == n-1) == 1
)
m.addConstrs(
    gp.quicksum(x[i, row['destination']] for index, row in df.iterrows() if row['origin'] == i) 
    - 
    gp.quicksum(x[row['origin'], i] for index, row in df.iterrows() if row['destination'] == i) 
    == 
    0 for i in range(1, n-1)
)
m.addConstrs(x[i, j] >= 0 for i in range(n) for j in range(n))
m.addConstrs(x[i, j] <= 1 for i in range(n) for j in range(n))
print('passed')

for rhs in m.getConstrs():
    print('%s: %g' % (rhs.ConstrName, rhs.pi))
```

```python
obj = gp.quicksum((df.iloc[i, 2] * x[df.iloc[i, 0], df.iloc[i, 1]]) for i in range(df.shape[0]))

m.setObjective(obj, GRB.MINIMIZE)
m.optimize()
```

```python
pathPrint = [str(z) for z in path]
pathPrint = ' -> '.join(pathPrint)
pathPrint
```

```python
length = 0
for each in range(len(path) - 1):
    for index, row in df.iterrows():
        if (row['origin'] == path[each]) and (row['destination'] == path[each + 1]):
            length += row['length']
length
```

```python
y = []
for i, rhs in enumerate(m.getConstrs()):
    y.append(rhs.pi)
    print('%s: %g' % ('y' + str(i), y[i]))
```

```python
dual = y[0] - y[99]
for i in range(101, 339):
    dual += y[i]
print('optimal dual objective value ', dual)
```
